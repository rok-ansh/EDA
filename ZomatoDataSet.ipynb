# To check Datatype

```python
final_df.dtypes
```

```python
# We can get the all the values related to country 
final_df.Country.value_counts()
```

```python
# If we want to chck what all country are we can use index
country_name = final_df.Country.value_counts().index
country_name
```

```python
# If we want only the values of we can use values
country_value=final_df.Country.value_counts().values
country_value
```

```python
# Lets create a pie chart - top 3 country  uses zomato


fig1, ax1 = plt.subplots()
ax1.pie(country_value[:3], labels=country_name[:3], autopct='%1.2f%%',
        shadow=True)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

### Observartion from Pie Chart : India serve around 94.39 out of total transaction, then US after that followed by UK

```python
final_df.columns
```

```python
# final_df.groupby(['Aggregate rating','Rating color','Rating text']).size()
# In order to make it dataframe use .reset_index()
# final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index()
# Now in we want to change the column name from 0 to some name we can use .rename(columns={0:'Rating Counts'})
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating Counts"})
ratings
```

Observation : 
1. Most of the people havent rated around 2k
2. 4.5 and above it was excellent 
3. 4.0 to 4.5 rating is very good
4. 3.5 to 3.9 rating is good and so on 

```python
# sns.barplot(data=ratings,x='Aggregate rating', y ='Rating Counts')
# Diagram looks small so we use matplot lib library to make it larger
import matplotlib
matplotlib.rcParams['figure.figsize']=(12,6) # This code is used to increase the size of figure 
sns.barplot(data=ratings,x='Aggregate rating', y ='Rating Counts')

```

```python
# Lets try to give color to the rating as given white green yellow orange , etc for that we can use hue='column_name'
sns.barplot(data=ratings,x='Aggregate rating', y ='Rating Counts', hue='Rating color')
```

```python
# One problem in the above chart is the color mapping is not proper we can we use palette['white','red',etc]
```

```python
sns.barplot(data=ratings,x='Aggregate rating', y ='Rating Counts', hue='Rating color', palette=['white','red','orange','yellow','green','green'])
```

 Observation:
 1. Maximum number of Rating is given between 2.5 to 3.4 which is avergae rating 


# Plot as count plot for Rating Color

```python
# We use this categorical variables
sns.countplot(x="Rating color", data=ratings,  palette=['blue','red','orange','yellow','green','green'])
```

```python
# Observation : White is only 1 records(frequency) its not based on Rating counts it based on no observation of Rating Color
# Orange records are the the most 10 times
```
